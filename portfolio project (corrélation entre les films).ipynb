```python
#visualiser la matrice de corrélation 
correlation_matrix=df.corr(method="pearson") 
sns.heatmap (correlation_matrix,annot=True)

plt.title("Matrice de corrélation ")
plt.xlabel ("caractéristiques des films")
plt.ylabel ("caractéristiques des films ")
plt.show()
```

```python
#company
#Assigner un numéro pour chaque compagnie 
df_numerized=df

for col_nom in df_numerized.columns:
    if (df_numerized[col_nom].dtype =="object"):
        df_numerized[col_nom]=df_numerized [col_nom].astype ("category")
        df_numerized[col_nom]=df_numerized [col_nom].cat.codes
df_numerized 
```

```python
#visualiser la matrice de corrélation 
correlation_matrix=df_numerized .corr(method="pearson") 
sns.heatmap (correlation_matrix,annot=True)

plt.title("Matrice de corrélation ")
plt.xlabel ("caractéristiques des films")
plt.ylabel ("caractéristiques des films ")
plt.show()
```
